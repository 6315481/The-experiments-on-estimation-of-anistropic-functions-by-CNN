```python
import matplotlib.pyplot as plt
import pickle
```

```python
with open('anistrophic_poly.pickle', 'rb') as f:
    results = pickle.load(f)
```

```python
dim_list = [100, 200, 500, 1000]
plt.ylim([0, 0.10])
for p in results.keys():
    plt.plot(dim_list, results[p], label='p = {}'.format(p))

plt.xlabel('dimension')
plt.ylabel('estimation error')
plt.legend(loc='best')
plt.show()
```
